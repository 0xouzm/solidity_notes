```python
import math

import matplotlib.pyplot as plt
import numpy as np

# Parameters
A = 1  # tokens minted per step
r = 0.01  # decay coefficient (small so balance doesn't go negative too fast)
B0 = 0  # initial balance
T = 1000  # number of steps
a = 0.01

# Initialize array
B = np.zeros(T + 1)
B[0] = B0


# Simulate recurrence
for t in range(T):
    B[t + 1] = B[t] + A - r * B[t]
    # prevent negative balances
    if B[t + 1] < 0:
        B[t + 1] = 0

# Plot
plt.figure(figsize=(10, 5))
plt.plot(range(T + 1), B, label="Balance B(t)")

plt.xlabel("Time step t")
plt.ylabel("Balance B(t)")
plt.grid(True)
plt.legend()
plt.show()
```
