```python
# Most significant bit and least significant bit
# Returns index of most significant bit
def msb(u256):
    assert 0 < u256 <= 2**256 - 1
    x = u256
    i = 0
    if x >= 2**128:
        x >>= 128
        i += 128
    if x >= 2**64:
        x >>= 64
        i += 64
    if x >= 2**32:
        x >>= 32
        i += 32
    if x >= 2**16:
        x >>= 16
        i += 16
    if x >= 2**8:
        x >>= 8
        i += 8
    if x >= 2**4:
        x >>= 4
        i += 4
    if x >= 2**2:
        x >>= 2
        i += 2
    if x >= 2:
        x >>= 1
        i += 1       
    return i

# Returns index of most significant bit (TODO)
def lsb(u256):
    assert 0 < u256 <= 2**256 - 1
    x = u256
    i = 0
    if x >= 2**128:
        x >>= 128
        i += 128
    if x >= 2**64:
        x >>= 64
        i += 64
    if x >= 2**32:
        x >>= 32
        i += 32
    if x >= 2**16:
        x >>= 16
        i += 16
    if x >= 2**8:
        x >>= 8
        i += 8
    if x >= 2**4:
        x >>= 4
        i += 4
    if x >= 2**2:
        x >>= 2
        i += 2
    if x >= 2:
        x >>= 1
        i += 1       
    return i
```
