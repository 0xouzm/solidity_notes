```python
# vault inflation attack
M = 1e6
total_supply = 0
user_shares = {}
total_assets = 0
users = [1, 2]

actions = [{
    "action": "deposit",
    "user": 1,
    "asset_amount": 1
}, {
    "action": "donate",
    "user": 1,
    "asset_amount": 100 * M
}, {
    "action": "deposit",
    "user": 2,
    "asset_amount": 100 * M
}]

def calc_shares(asset_amount):
    shares = 0
    if total_supply == 0:
        shares = asset_amount
    else:
        shares = asset_amount * total_supply // total_assets
    return shares

def calc_assets(shares):
    asset_amount = 0
    if total_supply > 0:
        asset_amount = shares * total_assets // total_supply
    return asset_amount

for action in actions:
    match action["action"]:
        case "deposit":
            user = action["user"]
            shares = calc_shares(action["asset_amount"])
            user_shares[user] = user_shares.get(user, 0) + shares
            total_supply += shares
            total_assets += action["asset_amount"]
        case "donate":
            total_assets += action["asset_amount"]
        case _:
            raise Exception()

    print(f'--- {action["action"]} ---')
    print(f'total supply: {total_supply / M}')
    print(f'total assets: {total_assets / M}')
    for user in users:
        shares = user_shares.get(user, 0)
        redeem_amount = calc_assets(shares)
        print(f'user {user}')
        print(f'shares: {shares / M}')
        print(f'redeem amount: {redeem_amount /  M}')
```

```python
# offset
# NOTE: works for M >= 1e14
M = 1e6
decimal_offset = 0
virtual_offset = 1

total_supply = 0
user_shares = {}
total_assets = 0
users = [1, 2]

actions = [{
    "action": "deposit",
    "user": 1,
    "asset_amount": 1
}, {
    "action": "donate",
    "user": 1,
    "asset_amount": 100 * M
}, {
    "action": "deposit",
    "user": 2,
    "asset_amount": 100 * M
}]

def calc_shares(asset_amount):
    return asset_amount * (total_supply + 10**decimal_offset) //  (total_assets + virtual_offset)

def calc_assets(shares):
    return shares * (total_assets + virtual_offset) // (total_supply + 10**decimal_offset)

for action in actions:
    match action["action"]:
        case "deposit":
            user = action["user"]
            shares = calc_shares(action["asset_amount"])
            user_shares[user] = user_shares.get(user, 0) + shares
            total_supply += shares
            total_assets += action["asset_amount"]
        case "donate":
            total_assets += action["asset_amount"]
        case _:
            raise Exception()

    print(f'--- {action["action"]} ---')
    print(f'total supply: {total_supply / M}')
    print(f'total assets: {total_assets / M}')
    for user in users:
        shares = user_shares.get(user, 0)
        redeem_amount = calc_assets(shares)
        print(f'user {user}')
        print(f'shares: {shares / M}')
        print(f'redeem amount: {redeem_amount /  M}')
```
